```python
import random

from PIL import Image
from math import sqrt
from IPython.display import display

bits = []
while len(bits) < 160000:
    bits.append(random.randint(0, 1))

width = int(sqrt(len(bits)))
height = int(sqrt(len(bits)))


image = Image.new('1', (width, height))

pixels = image.load()
for i in range(width):
    for j in range(height):
        pixels [i, j] = bits[j * width + i]

display(image)
```

```python
from PIL import ImageFilter
save_images = True

bits = []
width = 1920#int(sqrt(len(bits)))
height = 1080#int(sqrt(len(bits)))
while len(bits) < width*height:
    bits.append(((random.randint(0, 2**8-1)),(random.randint(0, 2**8-1)),(random.randint(0, 2**8-1))))

image = Image.new('RGB', (width, height))

pixels = image.load()
for i in range(width):
    for j in range(height):
        pixels[i, j] = bits[j * width + i]

display(image)
new_im = image.filter(ImageFilter.BLUR)
for i in range(2000):
    new_im = new_im.filter(ImageFilter.SHARPEN)
    new_im = new_im.filter(ImageFilter.BLUR)
    if save_images:
        # if i % 10 == 0:
        new_im.filter(ImageFilter.FIND_EDGES).save(f'output/im{i:04d}.png')
new_im = new_im.filter(ImageFilter.FIND_EDGES)

display(new_im)
```
